### 3-6 위의 전처리 과정을 전체 언어에 적용할 모듈 생성

```python
#비교 전
for raw_file in data_raw:
    raw_file.info()
```

```python
column_list = ["책제목", "작가", "가격", "할인가격", "ISBN", "출판사", "출판일", "link", "언어", "페이지수"]
```

```python
for raw_file in data_raw:
    del raw_file["Unnamed: 0"]
    raw_file['작가'] = raw_file['작가'].astype(str)
    raw_file['가격'] = raw_file['가격'].astype(str)
    raw_file['할인가격'] = raw_file['할인가격'].astype(str)
    raw_file['ISBN'] = raw_file['ISBN'].astype(str)
    raw_file['출판일'] = raw_file['출판일'].astype(str)
    raw_file['출판사'] = raw_file['출판사'].astype(str)
    raw_file.drop_duplicates(['ISBN'], inplace = True)
    raw_file.dropna(subset=[column for coloumn in column_list],inplace=True)
    raw_file.reset_index(drop=True, inplace = True)
```

```python
### 3-8 출판일 년도와 월로만 표시
result_list['출판일'] = result_list['출판일'].astype(str)
result_list['출판일'] = result_list['출판일'].str[:6]
```

```python
result_list['출판일'].unique()
```

```python
# 출판일 년도와 달 글자가 4 이상인 것만 추출
result_list = result_list[result_list['출판일'].str.len() >= 4]
```

```python
# 출판일에서 ".", "," 제거
for idx, row in result_list.iterrows():
    if "." in str(row["출판일"]):
        tmp = str(row["출판일"][:4] + "01")
        row["출판일"] = tmp
        result_list.iloc[idx] = row
    if "," in str(row["출판일"]):
        tmp = str(row["출판일"][:4] + "01")
        row["출판일"] = tmp
        result_list.iloc[idx] = row
```

```python
result_list['출판일'].unique()
```

```python
# 4자리수인 출판일 6자리로 변경
for idx, row in result_list.iterrows():
    if len(str(row["출판일"])) == 4:
        print(str(idx) + "   " + row["출판일"], end = "  ")
        tmp = str(row["출판일"] + "01")
        print(tmp)
        row["출판일"] = tmp
        result_list.iloc[idx] = row
```

```python
result_list['출판일'].unique()
```

```python
# 끝자리가 00인것 삭제 출판일 6자리로 변경
for idx, row in result_list.iterrows():
    if str(row["출판일"])[4:6] == "00":
        print(str(idx) + "   " + row["출판일"], end = "   ")
        tmp = str(row["출판일"][:4] + "01")
        print(tmp)
        row["출판일"] = tmp
        result_list.iloc[idx] = row
```

```python
result_list['출판일'].unique()
```

```python
# 4자리수인 출판일 6자리로 변경
for idx, row in result_list.iterrows():
    if (str(row["출판일"])).isdigit() == False:
        print(str(idx) + "   " + row["출판일"])
        tmp = np.nan        
        row["출판일"] = tmp
        result_list.iloc[idx] = row
```

```python
result_list['출판일'].unique()
```

```python
result_list = result_list[result_list["출판일"].isnull() == False]
```

```python
result_list["출판일"] = result_list["출판일"].astype(int)
```

```python
result_list
```

```python
### 3-9 ISBN 국내 출판 번호로 slicing
result_list["ISBN"] = result_list["ISBN"].str.split().str[1].str.slice(start = 0, stop = 5) + result_list["언어"][:]
```
